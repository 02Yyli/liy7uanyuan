```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0,2 * np.pi,200)
y = np.sin(x)


fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,2,3]
fig.ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('Ientry a')
ax.set_ylabel('entry b')
plt.show
```

```python
import matplotlib.pyplot as plt  
  
plt.rcParams['font.sans-serif']=['simHei']  # 设置字体  
plt.rcParams['axes.unicode_minus']=False  # 正常显示负号  
  
year = [2017, 2018, 2019, 2020]  
people = [20, 40, 60, 70]  
# 生成图表  
plt.scatter(year, people)  
plt.xlabel('年份')  
plt.ylabel('人口')  
plt.title('人口增长')  
  
# 设置填充选项：参数分别对应x坐标，y坐标，y坐标填充起始值，填充颜色  
plt.fill_betweenx(people, year, color='yellow')  
  
# 显示图表  
plt.show()
```
